```python
import pandas as pd
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
def parse_output(output):
    df = pd.DataFrame()
    for line in output:
        line = line.replace('\n', '')
        parsed_line = list(filter(lambda x: x.strip()!='', line.split(' ')))
        df = df.append({"time": parsed_line[0], 'port': parsed_line[1], 'value': float(parsed_line[2])}, ignore_index=True)
    return df

def verify_data_consistency(df1, df2):
    if (len(df1) != len(df2)):
        raise Exception('Check: {} == {}'.format(len(df1), len(df2)))
        
def process_data(df1, df2):
    df = pd.DataFrame()
    for i in range(0, len(df1)):
        row = {
            'time': df1.iloc[i]['time'],
            'obtained_energy_value': df1.iloc[i]['value'],
            'consumed_energy_value': df2.iloc[i]['value'],
            'net_energy_obtained':  df1.iloc[i]['value'] - df2.iloc[i]['value']
        }
        df = df.append(row, ignore_index = True)
    return df

def process_file(f):
    lines = f.readlines()

    df = parse_output(lines)

    df1 = df[df['port'] == 'obtained_energy']
    df2 = df[df['port'] == 'consumed_energy']

    df1= df1.set_index('time', drop=False)
    df2= df2.set_index('time', drop=False)

    verify_data_consistency(df1, df2)

    df3 = process_data(df1, df2)
    return df3
```

```python
import os

df = pd.DataFrame()
for file in os.listdir():
    if (not file.startswith('sim')):
        continue
    print(file)
    
    with open(file + '/output', 'r') as f:
        current_df = process_file(f)
        current_df['file'] = file
        df = df.append(current_df)

df
```

```python
fig = plt.figure()
print(fig)
plt.figure(figsize=(16, 6))
ax = sns.scatterplot(x='time', y='net_energy_obtained',data=df, hue='file', s=30)
ax.xaxis.set_major_locator(ticker.LinearLocator(30))
plt.xticks(rotation=45, ha='right')
plt.title("Net energy obtained for each 6 mins.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(5, 5))
df2 = df.groupby(['file']).apply(lambda x: x['net_energy_obtained'].sum()).to_frame()
df2['file'] = df2.index
df2.columns = ['net_energy_obtained', 'file']

ax = sns.barplot(x='file', y='net_energy_obtained',data=df2)
plt.title("Total net energy obtained")
```
